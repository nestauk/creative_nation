```python
#Installs

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

from cartopy.io import shapereader
from cartopy.feature import ShapelyFeature

file = "../data/external/ttwa_uk/Travel_to_Work_Areas_December_2011_Full_Extent_Boundaries_in_United_Kingdom.shp"
```

```python
ttwas = ShapelyFeature(shapereader.Reader(file).geometries(),
                      ccrs.PlateCarree(),edgecolor='black',facecolor='red')

```

```python
#Create the axes and use a projection

ax = plt.axes(projection=ccrs.GOOGLE_MERCATOR)

#Draw coastlines
ax.coastlines(resolution='50m')

#Set extent
#ax.set_extent([-12,3,49,60])

ax.add_feature(ttwas,facecolor='grey')

plt.show()
```
